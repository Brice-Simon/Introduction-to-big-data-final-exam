```python
import pandas as pd
```

```python
# Load the datasets
tourism_df = pd.read_csv('API_ST.INT.ARVL_DS2_en_csv_v2_22950.csv', skiprows=4)
air_transport_df = pd.read_csv('API_IS.AIR.PSGR_DS2_en_csv_v2_122328.csv', skiprows=4)
```

```python
# Filter data for Rwanda
air_transport_df_rwanda = air_transport_df[air_transport_df['Country Name'] == 'Rwanda']
tourism_df_rwanda = tourism_df[tourism_df['Country Name'] == 'Rwanda']

# Reshape the data using melt, dropping NA values
air_transport_df_rwanda = air_transport_df_rwanda.melt(id_vars='Country Code', var_name='Years', value_name='Air_transport').dropna().reset_index()
tourism_df_rwanda = tourism_df_rwanda.melt(id_vars='Country Code', var_name='Years', value_name='Tourism').dropna().reset_index()

# Remove unwanted rows and set index
air_transport_df_rwanda = air_transport_df_rwanda.drop(['index', 'Country Code'], axis=1)[3:]
tourism_df_rwanda = tourism_df_rwanda.drop(['index', 'Country Code'], axis=1)[3:]

air_transport_df_rwanda = air_transport_df_rwanda.set_index('Years')
tourism_df_rwanda = tourism_df_rwanda.set_index('Years')
```

```python
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.metrics import mean_squared_error, r2_score

print("Merging tourism and air transport datasets for Rwanda...")
combined_df_rwanda = tourism_df_rwanda.merge(air_transport_df_rwanda, how="inner", on="Years")
combined_df_rwanda['Tourism'] = combined_df_rwanda['Tourism'].astype(float)
combined_df_rwanda['Air_transport'] = combined_df_rwanda['Air_transport'].astype(float)

# Correlation test between normalized values
correlation, p_value = stats.pearsonr(combined_df_rwanda['Tourism'], combined_df_rwanda['Air_transport'])
print(f'Correlation: {correlation:.5f}, P-value: {p_value:.5f}')

# Linear regression between air transport and tourism
X = combined_df_rwanda['Air_transport'].astype(float)
y = combined_df_rwanda['Tourism'].astype(float)
# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(X, y)

# Function to predict values based on the regression line
def create_model(x):
    return slope * x + intercept

# Generate regression line values
model_line = list(map(create_model, X))

mean = mean_squared_error(model_line, y)
print(f"Mean Squared Error: {mean:.5f}")

r2 = r2_score(model_line, y)
print(f"Root Mean Squared Error: {r2:.5f}")

# Print model coefficients
print(f"Linear Regression Model: y = {intercept:.5f} + {slope:.5f} * x")

# Example Prediction for 2021
air_transport_2021 = float(air_transport_df_rwanda.loc['2021']['Air_transport'])
tourism_2021_predicted = create_model(air_transport_2021)
print(f"Predicted Tourism for 2021: {tourism_2021_predicted:.5f}")
```
